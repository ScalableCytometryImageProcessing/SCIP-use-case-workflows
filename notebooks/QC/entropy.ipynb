```python
import numpy
import scipy.stats
import seaborn
```

```python
data = numpy.random.normal(1, .5, size=1000)
noise = numpy.random.gamma(.2, 1, size=1000)
```

```python
seaborn.displot(x=noise)
```

```python
seaborn.stripplot(x=data+noise)
```

```python
scipy.stats.entropy(numpy.histogram(data)[0])
```

```python
scipy.stats.entropy(numpy.histogram(data+noise)[0])
```
